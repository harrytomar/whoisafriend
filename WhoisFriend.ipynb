```python
dftest=pd.read_csv('/kaggle/input/whoisafriend/test.csv')
dftrain=pd.read_csv('/kaggle/input/whoisafriend/train.csv')
```

```python
dftrain.head(2)
```

```python
dftrain=dftrain.drop(['ID','Person A','Person B','Interaction Type','Moon Phase During Interaction'],axis=1)
```

```python
dftrain.head(1)
```

```python
X=dftrain.iloc[:,:-1]
Y=dftrain.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
regressor=LogisticRegression()
regressor.fit(x_train,y_train)
```

```python
pred=regressor.predict(x_test)
```

```python
print(pred)
```

```python
from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,pred.round())
```

```python
score.shape
```

```python
print(score)
```
